```python
pd.concat([df1, df2, df3], axis=0, join='outer')
```

```python
pd.concat([df1, df3], axis=0, join='inner')
```

```python
person = pd.read_csv('data/data/survey_person.csv')
site = pd.read_csv('data/data/survey_site.csv')
survey = pd.read_csv('data/data/survey_survey.csv')
visited = pd.read_csv('data/data/survey_visited.csv')
```

```python
v_subset = visited.loc[[0, 2, 6]]
```

```python
v_subset
```

```python
pd.merge()
site.merge()
# 양쪽 컬럼이름이 같을 땐 on쓰면 된다.
# 양쪽 컬럼 이름이 다를 때
# suffixes 
```

```python
# 왼쪽엔 name 컬럼, 오른쪽엔 site 컬럼
# 기본적으로 조인은 inner join
site.merge(v_subset, left_on='name', right_on='site')
```

```python
site.merge(visited, left_on='name', right_on='site')
```

```python
person.merge(survey, left_on='ident', right_on='person')
```

```python
ps = person.merge(survey, left_on='ident', right_on='person')
```

```python
vs = visited.merge(survey, left_on='ident', right_on='taken')
```

```python
ps.head(2)
```

```python
vs.head(2)
```

```python
ps.merge(vs, 
         left_on=['ident', 'taken', 'quant', 'reading'], 
         right_on=['person', 'ident', 'quant', 'reading'], 
         how='outer')
```
