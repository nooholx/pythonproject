```python
tips = sns.load_dataset('tips')
```

```python
tips.head(2)
```

```python
tips.info()
```

```python
# 히스토그램은 분포와 빈도를 살펴볼 때 쓴다
# 데이터의 연속적인 값은 모든 값이 하나씩밖에 존재하지 않음. 범위안에 데이터가 몇건이 있다. 이런식으로 분류됨
# 그 빈도를 나타낸게 히스토그램
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 1, 1)
# 보통 bins 기본값은 10으로 되어있다. 
ax1.hist(tips['total_bill'], bins=10, density=True)
ax1.set_title('Histogram of Total Bill')
ax1.set_xlabel('Frequency')
ax1.set_ylabel('total bill')

# 그래프 여러 개를 한꺼번에 그렸을 때 마지막이 안찍히는 경우가 있는데 그때 해주면 좋다.
plt.show()
```

```python
print(tips[['total_bill', 'tip']].corr())
scatter_plot = plt.figure()
# 111안에 컴마는 안넣어도 된다.
ax1 = scatter_plot.add_subplot(111)
ax1.scatter(tips['total_bill'], tips['tip'])
plt.show()
```

```python
boxplot = plt.figure()
# 1행 1열에 1개짜리로 만들겠다 
ax1 = boxplot.add_subplot(111)
data = [tips[tips['sex'] =='Female']['tip'], tips[tips['sex'] =='Male']['tip']]
ax1.boxplot(data, labels=['Female', 'Male'])
plt.show()
```

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
tips['color'] = tips['sex'].apply(recode_sex)
```

```python
tips
```

```python
scatter_plot = plt.figure()
```

```python
ax1 = scatter_plot.add_subplot(111)
```

```python
ax1.scatter(tips['total_bill'],
           tips['tip'],
           s=tips['size']*10,
           c=tips['color'],
           alpha=0.5)
ax1.set_title('산정도')
```

```python
scatter_plot = plt.figure()
ax1 = scatter_plot.add_subplot(1,1,1)
ax1.scatter(tips['total_bill'],tips['tip'], s=tips['size']*10, c=tips['color'],alpha=0.5 )
plt.show()
```
