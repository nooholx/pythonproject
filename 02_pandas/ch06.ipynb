```python
ebola.info()
```

```python
# nan이면 true 
# sum 1번 -> 컬럼단위
# sum 2번 ->
ebola.isna().sum().sum()
```

```python
ebola.fillna(0)
```

```python
# 특정 컬럼만 채우고 싶을 떄
ebola['Cases_Liberia'].fillna(0)
```

```python
# inplace 원본 그대로 리턴값은 적용해서 출력해주는 것
ebola.fillna(0).head(2)
```

```python
ebola.head(5)
```

```python
# f가 붙은 것은 앞에 것을 채워라
ebola.fillna(method = 'ffill')
```

```python
# 뒷쪽에 있는 값으로 채워라
ebola.fillna(method = 'bfill')
```

```python
# 앞 뒤의 중간값으로 채워라
ebola.interpolate()
```

```python
ebola['Cases_Guinea'].fillna(ebola['Cases_Guinea'].mean())
```
