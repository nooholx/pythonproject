```python
# melt를 이용하여 월별데이터를 '기간' 컬럼으로 가격은 '평당분양가격'으로 처리하여 first_df_melt로 저장
first_df_melt = first_df.melt(id_vars='지역', var_name = '기간', value_name = '평당분양가격')
first_df_melt.head(3)
```

```python
# 기간 컬럼을 연도, 월 컬럼으로 분리
first_df_melt['연도'] = first_df_melt['기간'].str.split('년').str[0].astype(int)
first_df_melt['월'] = first_df_melt['기간'].str.split('년').str[1].str.replace('월', '').astype(int)
```

```python
first_df_melt.head(3)
```

```python
first_df_melt.info()
```

```python
# last_df 전용면적에서 전체인 데이터만 last_df_prepare로 저장
last_df_prepare = last_df[last_df['전용면적'] == '전체'].copy()
```

```python
first_df_melt.head(2)
```

```python
# 지역명 -> 지역 컬럼으로 변경
last_df_prepare.columns = ['지역', '연도', '월', '분양가격', '평당분양가격', '전용면적']
# 이렇게 해도 되지만 컬럼이 엄청 많을 때는 비효율적 
```

```python
# rename을 사용
# 값을 딕셔너리로 넣는다.
last_df_prepare.rename(columns={'지역명' : '지역'}, inplace = True)
```
