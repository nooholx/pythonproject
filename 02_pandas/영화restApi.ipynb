```python
# 초를 받아서 날짜형식으로 변환해줌. ->  strftime으로 시간값으로부터 문자열로 바꿔줌
def cineBoxInfo():
    movieDate = time.strftime('%Y%m%d', time.localtime(time.time()))
    cine = []
    for i in range(30):
        # 어제 날짜가 필요하니 오늘 날짜 -1 해야함. 
        # 그러려면 movieDate의 형식으로는 구할 수 없지.
        # 문자열을 날짜형식으로 바꿔주는 작업(movieDate문자열이 '%Y%m%d'형식이다. 이것을 날짜 형식으로 바꿔라)
        # strptime은 문자열에서 시간값으로 바꾸는거
        date_obj = datetime.datetime.strptime(movieDate, '%Y%m%d') 
        # 하루 빼기 (전일 날짜 구하기 : 시간)
        # 결과 : datetime.datetime(2023, 4, 09, 0, 0)
        date_obj_temp = date_obj - datetime.timedelta(days=1)
        # 전일 날짜를 문자열로 만들기
        movieDate = date_obj_temp.strftime('%Y%m%d')
        print(movieDate, end=' ')
        url = f'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key={key}&targetDt={movieDate}'
        r = urllib.request.urlopen(url)
        data = json.loads(r.read())
        result = data['boxOfficeResult']['dailyBoxOfficeList']
        # result값을 dataframe값으로 만들기
        for i in range(len(result)):
            # 데이터를 수집한 날짜를 넣어주기
            result[i]['targetDt'] = movieDate
            cine.append(result[i])
    
    # 데이터 만들기
    df = pd.DataFrame(cine)
    
    # 데이터 (df) 를 csv로 저장하기 
    df.to_csv('cine.csv', index=False)
    
    return df
```
