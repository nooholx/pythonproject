# CGV

```python
from urllib import request
from bs4 import BeautifulSoup
```

```python
# 1. urllib을 사용하는 방법
url = 'http://www.cgv.co.kr/movies'
response = request.urlopen(url)
# print(request.urlopen(url).read().decode('utf8'))
soup = BeautifulSoup(response, 'lxml')
# print(soup)
```

```python
result = soup.select('strong.title')
# print(result)

movie_title = []
for item in result:
    # print(item.get_text())
    movie_title.append(item.get_text())
print(movie_title)
```

```python
# 2. requests를 사용하는 방법
import requests

url = 'http://www.cgv.co.kr/movies'
response = requests.get(url)
response.text
result = soup.select('strong.title')
# print(result)

movie_title = []
for item in result:
    # print(item.get_text())
    movie_title.append(item.get_text())
print(movie_title)
```

# Bugs Music

```python
from urllib import request
from bs4 import BeautifulSoup
```

```python
url = 'https://music.bugs.co.kr/chart'
```

```python
r = request.urlopen(url)
soup = BeautifulSoup(r, 'lxml')
# print(soup)
title_list = []
artist_list = []
```

```python
title = soup.select('p.title > a')
print(title)
```

```python
for item in title:
    # print(item.string)
    # print(item.text)
    # print(item.get_text())
    title_list.append(item.get_text())
print(title_list)
```

```python
artist = soup.select('p.artist')

artist
```

```python
for item in artist:
    # print(item.string)
    artist_list.append(item.get_text().strip())

artist_list
```

```python
print(len(title_list), len(artist_list))
```

```python
result = list(zip(title_list, artist_list))
```
