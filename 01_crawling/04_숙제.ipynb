```python
import requests
from bs4 import BeautifulSoup
```

```python
# http://www.cgv.co.kr/movies/?lt=1&ft=0 
#무비차트 영화 제목 크롤링
url = 'http://www.cgv.co.kr/movies/?lt=1&ft=0'
r = requests.get(url).text
# print(r)
soup = BeautifulSoup(r, 'lxml')
# print(soup)
movie_name = [] #영화명
# price = [] #가격

data = soup.select('div.box-contents > a') #div 아래있는 a태그
# print(data)

# # get_text().strip()과 string.strip()결과가 같다. 이렇게 써도 되는 듯
for item in data :
    temp = item.get_text().strip()
    movie_name.append(temp)
print(movie_name)
```

```python
url = 'https://music.bugs.co.kr/chart'
r = requests.get(url).text
soup = BeautifulSoup(r, 'lxml')

song_title = []
artist = []

# 곡명 가져오기
data = soup.select('p.title > a')
for item in data:
    song_title.append(item.string.strip())

# 아티스트 가져오기
data = soup.select('p.artist > a')
for item in data:
    temp = item.string.strip()
    # 피처링 아티스트가 아닌 경우만 추가
    if item.select_one('span') is None:
        artist.append(temp)

print(len(song_title), len(artist))
```
