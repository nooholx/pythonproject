## 03. 검색어 넣어서 html 페이지 가져오기
### 01. urllib이용

```python
from urllib import request, parse
```

```python
# ?를 기준으로 앞은 주소 뒤는 쿼리스트링
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query='
```

```python
query = input('검색할 키 >>>')
# ?를 기준으로 앞은 주소 뒤는 쿼리스트링
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query='+query
print(url)
# 검색할 때 한글 그대로 쓰면 오류가 남. utf8 인코딩 해야해.
request.urlopen(url).read()
```

```python
url = 'https://search.naver.com/search.naver?'
query = input('검색할 키 >>>')
values = {
    # 'where':'nexearch', 
    # 'sm':'top_hty',
    # 'fbm':1,
    # 'ie':'utf8',
#     query는 필수로 넣어야함.
    'query': query
}

# 결과값을 문자열로 return해줌(utf8로 인코딩되어서 들어감)
params = parse.urlencode(values)
# print(params)
url = url + params
print(url)

request.urlopen(url).read().decode('utf8')
```

### 02. requests 이용

```python
import requests
```

```python
query = input('검색할 키 >>> ')
url = 'https://search.naver.com/search.naver?'
values = {
    # 'where':'nexearch', 
    # 'sm':'top_hty',
    # 'fbm':1,
    # 'ie':'utf8',
#     query는 필수로 넣어야함.
    'query': query
}
r = requests.get(url, params=values)
# print(r)
# 읽어들이기
r.text
```
