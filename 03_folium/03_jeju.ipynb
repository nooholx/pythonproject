```python
# 1. 주소 정보를 이용하여 위도, 경도 값을 저장합니다.
#    '위치' 컬럼에 (위도, 경도) 형태로 저장
#    주소를 찾아서 저장하는 함수를 작성하여 처리
#    함수 이름은 get_coder(query) 리턴값 : (위도, 경도) 만일 못찾았다면 '결과없음'
#    함수사용, apply()를 이용하여 데이터 프라임에 적용하시오.
```

```python
import requests
```

```python
REST_API_KEY = '96639ee0a233e38fabb5e2b0b8b98f3d'
```

```python
def geo_coder(query):
    code = 'PO3'
    # query.split('(')  === > ['제주특별자치도 제주시 중앙로7길 15 ', '일도일동)']
    # query.split('(')[0]  === > 제주특별자치도 제주시 중앙로7길 15 
    # split을 쓰면 리스트형태로 반환되니까
    query = query.split('(')[0]
    url = f'https://dapi.kakao.com/v2/local/search/keyword.json?&query={query}&category_group_code={code}'
    header = {'Authorization': f'KakaoAK {REST_API_KEY}'}
    r = requests.get(url, headers=header)
    jsondata = r.json()
    try :
        return jsondata['documents'][0]['y'], jsondata['documents'][0]['x']
    except:
        return '결과없음'

df['위치'] = df['주소'].apply(geo_coder)
```

```python
# 2. '위치' 컬럼의 값을 Marker위치값으로, tooltip으로 읍면동을 사용하여 지도에 표시(클러스터 이용)
# 함수사용, apply()를 사용용
```

```python
import folium
from folium.plugins import MarkerCluster
```

```python
def display(data, map):
    # print(data[0], data[1])
    if data[0] != '결과없음':
        folium.Marker(data[0], tooltip=data[1]).add_to(map)
       
map = folium.Map((33.3616666, 126.5291666), zoom_start=9)
map_c = MarkerCluster().add_to(map)

df[['위치', '읍면동']].apply(display, map=map_c, axis=1)

map
```

```python
map = folium.Map((33.3616666, 126.5291666), zoom_start=9)
map_c = MarkerCluster().add_to(map)

for index, row in df.iterrows():
    # print(index, row)
    # print('------------')
    if row['위치'] != '결과없음':
        folium.Marker(row['위치'], tooltip=row['읍면동']).add_to(map_c)
map
```

```python
# 3. '일반현황' 컬럼의 인구수만 분리하여 '인구수'라는 int형 컬럼을 생성
```

```python
df['인구수'] = df['일반현황'].str.split('/').str[0].str.split('(').str[0].str.split().str[-1].str.replace('명','').astype(int)
```

```python
import re

df['일반현황'].apply(lambda x:int(re.findall('[0-9]{4,}', x)[-1]))
```
